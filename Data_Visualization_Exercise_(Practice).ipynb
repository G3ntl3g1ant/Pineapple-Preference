# Pinapple Preference

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pinapple on Pizza data frame
filename ='https://docs.google.com/spreadsheets/d/e/2PACX-1vSSCXERaxwu8jhbJ165RrWytNpEfhA4gE6hUa5e07YiiCSmU4kHuwRZ48loaaGY0Q/pub?gid=1344951467&single=true&output=csv'
df = pd.read_csv(filename)
df.head()
```

```python
# Going over intial look at the information for this Data Frame
df.info()
```

```python
# Checking for missing data 
df.isna().sum()
```

```python
# Male preffered toppings
sns.barplot(data=df, y= 'Topping', x='Male');
```

From this information we can infer that it looks to be that Males specifically like the more Protein/Meat Pizza Toppings with Ham and Chicken at close 1st choices, Mushrooms and Peppers at close 2nd being vegitarian options.

```python
# Female preffered toppings
sns.barplot(data=df, y= 'Topping', x='Female');
```

While in comparison, Women seem to prefer the more vegitarion option, with Mushrooms, Onions and Peppers as the top 3 choices.

```python
# Creating values for the following graphs 
male = df['Male'].values
female = df['Female'].values
total = df['Total'].values
topping = df['Topping'].values
```

```python
# Breakdown of Male and Female tastes in pizza toppings
plt.scatter(female, topping, c = 'r', label = 'Females')
plt.scatter(male, topping, c = 'b', label = 'Males')
plt.legend(loc=(1.02, 0));
plt.title('Tastes in Pizza Toppings')
plt.xlabel('Count of Males and Females')
plt.ylabel('Pizza Toppings')
plt.grid()
```

Here we can see that the information above essentially reiterates the information from the prior 2 barplots. Women seem to prefer the veggitarian option while Men seem to prefer the carniverous options.

But when it comes to pineapple, it seems that it is towards the 50% mark rather than the cliche' beleif that it is an undesired topping.

```python
# Putting everything together to get an overall understanding
plt.scatter(total, topping, c = 'k', label = 'Percentages')
plt.plot(total, topping, c = 'k', label = 'Percentage Totals of Both Sexes')
plt.scatter(female, topping, c = 'r', label = 'Females')
plt.scatter(male, topping, c = 'b', label = 'Males')
plt.legend(loc=(1.02, 0));
plt.title('Tastes in Pizza Toppings')
plt.xlabel('# of Participants')
plt.ylabel('Pizza Toppings')
plt.grid()
```

Using the overall total percentages of people when it comes to preferred toppings, pineapple comes out at 42%. With more women preferring it to men. 

But much more than was previously believed, maybe there will be a growth preference for pineapple on Pizza.
